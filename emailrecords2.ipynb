```python
def collatedata(path):
    cols = ['#', 'AB', 'Date','Time Stamp','Pitch','Strike Zone', 'P. Type', 'Velo', 'LA', 'Dist', 'Res', 'Type', 'Horiz. Angle',
    'Pts', 'Hand Speed', 'BV', 'Trigger to Impact', 'AA', 'Impact Momentum', 'Strike Zone Bottom', 'Strike Zone Top',
    'Strike Zone Width','Vertical Distance','Horizontal Distance', 'X', 'Y', 'Z', 'Bat Material','User',
    'Pitch Angle', 'Batting', 'Level', 'Opposing Player', 'Tag']
    mypath = "c:\\tmp\\msgs"
    dlist = []
    onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    for fname in onlyfiles:
        fname = mypath+"\\"+fname
        #print(fname)
        try:
            dfnew = pd.read_csv(fname, sep='\s*,\s*', engine='python', header=0, names = cols)
        except:
            print(fname, 'reading error')
        else:
            cols = dfnew.columns
            if len(cols) < 30:
                print(fname, 'bogus file ignored')
            else:
                dlist.append(dfnew)
            #print(dfnew.User.unique(), dfnew.shape)
            #print(dfnew.columns)
            
    df = pd.concat(dlist, ignore_index=True)
    #print(fname, df.shape, dfnew.shape)
            
    return df
```
